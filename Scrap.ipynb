```r
install.packages("academictwitteR")
```

```r
library(academictwitteR)
library(tidyverse)
```

```r
set_bearer()
```

```r
library(yaml)
keys <- read_yaml("keys.yaml")
bearer <- keys$bearer
query_amlo <- build_query(query = "AMLO", is_retweet = FALSE)
```

```r
#Prueba
tweets1 <-
  get_all_tweets(
    query = query_amlo,
    start_tweets = "2022-03-29T00:00:00Z",
    end_tweets = "2022-04-19T00:00:00Z",
    data_path = "JSON/",
    bind_tweets = TRUE,
    bearer_token = bearer,
    n= 100
  )
```

```r
tweets1 <- tweets1 |> as_tibble() |> select(author_id, lang, text, created_at)
head(tweets1, 10)
```

```r
tweets2 <- get_all_tweets(query = query_amlo, 
                          start_tweets = "2022-03-28T00:00:00Z",
                          end_tweets = "2022-04-03T00:00:00Z",
                          data_path = "JSON/",
                          bind_tweets = TRUE,
                          bearer_token = bearer,
                          n = 170000)

tweets2 <- tweets2 |> as_tibble() |> select(author_id, lang, text, created_at)
head(tweets2, 10)
```

```r
write.csv(tweets2, "csvs/tweets_1rasem.csv")
```

```r
tweets3 <- get_all_tweets(query = query_amlo, 
                          start_tweets = "2022-04-04T00:00:00Z",
                          end_tweets = "2022-04-10T00:00:00Z",
                          data_path = "JSON/",
                          bind_tweets = TRUE,
                          bearer_token = bearer,
                          n = 170000)

tweets3 <- tweets3 |> as_tibble() |> select(author_id, lang, text, created_at)
head(tweets3, 10)
```

```r
write.csv(tweets3, "csvs/tweets_2rasem.csv")
```

```r
tweets4 <- get_all_tweets(query = query_amlo, 
                          start_tweets = "2022-04-11T00:00:00Z",
                          end_tweets = "2022-04-19T00:00:00Z",
                          data_path = "JSON/",
                          bind_tweets = TRUE,
                          bearer_token = bearer,
                          n = 180000)

tweets4 <- tweets4 |> as_tibble() |> select(author_id, lang, text, created_at)
head(tweets4, 10)
```

```r
write.csv(tweets4, "csvs/tweets_3rasem.csv")
```

```r
# Unimos la base:

tweets_juntos <- rbind(tweets2, tweets3, tweets4)
```

```r
# Hacemos un poco de limpieza:
tweets_juntos <- tweets_juntos |> mutate(row_num = row_number())
loc <- as_tibble(str_locate(tweets_juntos$text, "https://")[,1]) |> mutate(row_num = row_number())

# Eliminamos los urls:

tweets_juntos <- tweets_juntos |> left_join(loc, by = "row_num") |> mutate(value = if_else(is.na(value) == T, -1, value-1)) |> 
mutate(text2 = str_sub(text, 1, value))
```

```r
head(tweets_juntos, 5)
```

```r
tweets_juntos <- tweets_juntos |>  
  filter(lang == "es") 

write.csv(tweets_juntos, "csvs/tweets_todos.csv")
```

```r
# Recortamos la muestra aleatoriamente: 4800 tuits
tweets_sampled <- tweets_juntos |> slice_sample(n=4800)
write.csv(tweets_sampled, "csvs/tweets_sampled.csv")
```

```r
tweets2_siha <- tweets_sampled |> slice(1:800)
tweets2_mex <- tweets_sampled |> slice(801:1600)
tweets2_mike <- tweets_sampled |> slice(1601:2400)
tweets2_alvaro <- tweets_sampled |> slice(2401:3200)
tweets2_sergio <- tweets_sampled |> slice(3201:4000)
tweets2_zago <- tweets_sampled |> slice(4001:4800)
```

```r
write.csv(tweets2_siha, "csvs/tweets2_siha.csv")
write.csv(tweets2_mex, "csvs/tweets2_mex.csv")
write.csv(tweets2_mike, "csvs/tweets2_mike.csv")
write.csv(tweets2_alvaro, "csvs/tweets2_alvaro.csv")
write.csv(tweets2_sergio, "csvs/tweets2_sergio.csv")
write.csv(tweets2_zago, "csvs/tweets2_zago.csv")
```
